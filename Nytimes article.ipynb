```python
from bs4 import BeautifulSoup as BS
import requests
```

```python
r = requests.get("https://www.nytimes.com/interactive/2018/us/elections/calendar-primary-results.html")
# r stores the values from the link
```

```python
print(r.text[2000:3000]) #prints the first 999 words of the source 
```

```python
soup = BS(r.text, 'html.parser')
```

```python
#storing all the 
results = soup.find_all('div', attrs = {'class':'g-calendar-race-title'})
results2 =soup.find_all('div', attrs = {'class':'g-calendar-race-desc'})
results3 = soup.find_all('span', attrs = {'class':'g-key-label'})
results4 = soup.find_all('span',attrs = {'class':'g-key-d'})
```

```python
#building the dataset
Record = []
for result in results :
    Name = result.contents[0]
    Record.append(Name)
Record2 = []
for result2 in results2 :
    Reason = result2.contents[0].find('a')
    Record2.append(Reason)
Record3 = []
for result3 in results3 :
    Class = result3.contents[0]
    Record3.append(Class)
Record4 = []
for result4 in results4 :
    Candidate = result4.contents[0]
    Record4.append(Candidate)
```

```python
Record2
```

```python
import pandas as pd
```

```python
len(Record4)
```

```python
my_dict = dict(Name = Record, Reason = Record2, Class = Record3, Candidate = Record4)
```

```python
Elections_2018 = pd.DataFrame.from_dict(my_dict, orient='index')
```

```python
Elections_2018 = Elections_2018.transpose()
```
